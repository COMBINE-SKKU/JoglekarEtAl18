```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
dfEstrong = pd.read_csv('ExcitatoryStrong_0.r.dat', sep="\t", header=None)
dfEweak = pd.read_csv('ExcitatoryWeak_0.r.dat', sep="\t", header=None)

# Figure
fig,ax=plt.subplots()
ax.plot(dfEweak[0],dfEweak[1], 'g')
ax.plot(dfEstrong[0],dfEstrong[1],'m')
ax.set_ylabel('Excitatory rate (Hz)', fontsize='x-large')
ax.set_xlabel('Time (s)', fontsize='x-large')
ax.set_ylim([0,6])
ax.set_xlim([0,0.6])
ax.set_yticks([0,2,4,6])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.legend(['weak LBA','strong LBA'],prop={'size': 15},frameon=False)
fig.set_size_inches(10,5)
```
